# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Criando dados simples

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

# Criando um gr치fico b치sico

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Seno de x', color='b')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gr치fico B치sico com Matplotlib')
plt.legend()
plt.grid()
plt.show()
```
