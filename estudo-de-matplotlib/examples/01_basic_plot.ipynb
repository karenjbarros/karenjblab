# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Criando dados simples

Ao utilizar o `np.linspace`estamos gerando 100 valores entre 0 e 10, o que utilizaremos para popular o eixo X. 

Por ser um código de `numpy`, nós estamos gerando um array, que é uma estrutura otimizada para cálculos numéricos, mais rápida e eficiente em operações vetoriais.

```python
x = np.linspace(0, 10, 100)
print(type(x))
print(x)
```

# Explicando o Seno

O seno é uma função matemática fundamental na trigonometria. Ele descreve a relação entre os lados de um triângulo retângulo e também é amplamente utilizado para representar movimentos ondulatórios, como ondas sonoras e eletromagnéticas. 

Através do uso do `np.sin(x)`, aplicaremos a função seno a cada elemento do array x, criando a base de y, que será usado para plotar o gráfico.

A função seno, definida em todos os números reais, sempre assume valores entre -1 e 1. Essa característica decorre tanto da sua definição geométrica usando o círculo unitário quanto da relação trigonométrica fundamental.

```python
y = np.sin(x)
print(y)
```

# Criando um gráfico básico

##  Faremos um gráfico de Curva Senoidal

O termo "senoidal" deriva diretamente da função seno, cuja representação gráfica consiste em uma onda contínua que varia entre -1 e 1 e se repete a cada 2π. Essa propriedade intrínseca determina o comportamento ondulatório da curva.


Historicamente, essa nomenclatura foi adotada em matemática e física para descrever fenômenos oscilatórios, como movimentos harmônicos simples e sinais de corrente alternada.

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Seno de x', color='b')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Curva Senoidal')
plt.legend()
plt.grid()
plt.show()
```
