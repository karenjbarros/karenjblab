# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Definindo uma semente para reprodutibilidade

A função `np.random.seed(42)` é usada em Python, com a biblioteca NumPy, para fixar o estado inicial do gerador de números aleatórios.

Isso significa que, ao definir uma "semente" (no caso, o número 42), os números aleatórios gerados posteriormente serão sempre os mesmos, garantindo reprodutibilidade.

```python
np.random.seed(42)
```

# Gerando 1000 pontos de dados com distribuição normal

A função `np.random.randn(1000)` do NumPy gera 1.000 números aleatórios seguindo uma distribuição normal padrão (ou distribuição Gaussiana), com as seguintes características:

Média (mean): 0

Desvio padrão (standard deviation): 1

Distribuição: Os números são distribuídos de forma simétrica em torno da média, com a maioria dos valores concentrados próximos a 0, e alguns valores mais distantes (cauda da distribuição).

```python
data = np.random.randn(1000)

data
```

# Criando um gráfico

## Seaborn

O Seaborn é uma biblioteca de visualização de dados em Python baseada no Matplotlib. 

Ele é amplamente utilizado para criar gráficos estatísticos de forma simples, estilizada e informativa. 

O Seaborn facilita a análise exploratória de dados, integrando-se bem com estruturas como os DataFrames do Pandas e oferecendo suporte a gráficos avançados, como histogramas, gráficos de densidade, mapas de calor e muito mais.

## Curva KDE

A Curva KDE (Kernel Density Estimate) é uma representação gráfica gerada a partir da técnica de Estimativa de Densidade por Kernel (KDE), que é usada para estimar a função de densidade de probabilidade (FDP) de uma variável aleatória. 

Essa técnica é uma abordagem não-paramétrica, ou seja, não assume um modelo específico para os dados, e é amplamente utilizada para visualizar distribuições de dados de forma suave e contínua.



```python
# Criando uma figura para o histograma com KDE
plt.figure(figsize=(8, 5))

# Plotando o histograma com KDE usando Seaborn
sns.histplot(data, kde=True, bins=30, color='purple', alpha=0.7)

# Personalizando o gráfico
plt.xlabel('Valores', fontsize=12, fontweight='bold')
plt.ylabel('Frequência', fontsize=12, fontweight='bold')
plt.title('Histograma com KDE usando Seaborn', fontsize=14, fontweight='bold')
plt.grid(True, linestyle=':', linewidth=0.7)

# Exibindo o gráfico
plt.show()
```
